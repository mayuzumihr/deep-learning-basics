```python
import MeCab
import ipadic
```

```python
tagger = MeCab.Tagger()
text = "吾輩は猫である。名前はまだない。"
result = tagger.parse(text)
print(result)

tagger = MeCab.Tagger(f'-Odump')
text = "吾輩は猫である。名前はまだない。"
result = tagger.parse(text)
print(result)

tagger = MeCab.Tagger(f'-Owakati')
text = "吾輩は猫である。名前はまだない。"
result = tagger.parse(text)
print(result)
```

```python
print(ipadic.DICDIR)

tagger = MeCab.Tagger(f'-d "{ipadic.DICDIR}"')
text = "吾輩は猫である。名前はまだない。"
result = tagger.parse(text)
print(result)

tagger = MeCab.Tagger(f'-Odump -d "{ipadic.DICDIR}"')
text = "吾輩は猫である。名前はまだない。"
result = tagger.parse(text)
print(result)

tagger = MeCab.Tagger(f'-Owakati -d "{ipadic.DICDIR}"')
text = "吾輩は猫である。名前はまだない。"
result = tagger.parse(text)
print(result)

tagger = MeCab.Tagger(f'-Ochasen -d "{ipadic.DICDIR}"')
text = "吾輩は猫である。名前はまだない。"
result = tagger.parse(text)
print(result)

tagger = MeCab.Tagger(f'-Oyomi -d "{ipadic.DICDIR}"')
text = "吾輩は猫である。名前はまだない。"
result = tagger.parse(text)
print(result)
```
