### ■ パーセプトロン

- もっとも簡単なパーセプトロンを考えてみる
- w, x ともに個別の変数として扱う
- 【参考】閾値が theta。意味としては bias 項と同じ

```python
import numpy as np
import matplotlib.pyplot as plt
```

### ■ AND ゲート

```python
#------------------------------------------------------------------------------
# AND Gate
#------------------------------------------------------------------------------
def _and_(x1, x2):

  w1 = 0.5
  w2 = 0.5
  b = -0.75

  ans = (w1 * x1) + (w2 * x2) + b

  if ans > 0:
    return 1
  else:
    return 0

print(f'AND(0, 0) : {_and_(0, 0)}')
print(f'AND(0, 1) : {_and_(0, 1)}')
print(f'AND(1, 0) : {_and_(1, 0)}')
print(f'AND(1, 1) : {_and_(1, 1)}')
```

### ■ OR ゲート

```python
#------------------------------------------------------------------------------
# OR Gate
#------------------------------------------------------------------------------
def _or_(x1, x2):

  w1 = 0.5
  w2 = 0.5
  b = -0.25

  ans = (w1 * x1) + (w2 * x2) + b

  if ans > 0:
    return 1
  else:
    return 0

print(f'OR(0, 0) : {_or_(0, 0)}')
print(f'OR(0, 1) : {_or_(0, 1)}')
print(f'OR(1, 0) : {_or_(1, 0)}')
print(f'OR(1, 1) : {_or_(1, 1)}')
```

### ■ NAND ゲート

```python
#------------------------------------------------------------------------------
# NAND Gate
#------------------------------------------------------------------------------
def _nand_(x1, x2):

  w1 = -0.5
  w2 = -0.5
  b = 0.75

  ans = (w1 * x1) + (w2 * x2) + b

  if ans > 0:
    return 1
  else:
    return 0

print(f'NAND(0, 0) : {_nand_(0, 0)}')
print(f'NAND(0, 1) : {_nand_(0, 1)}')
print(f'NAND(1, 0) : {_nand_(1, 0)}')
print(f'NAND(1, 1) : {_nand_(1, 1)}')
```

### ■ XOR ゲート

```python
#------------------------------------------------------------------------------
# XOR Gate
#------------------------------------------------------------------------------
def _xor_(x1, x2):
  return _and_(_nand_(x1, x2), _or_(x1, x2))

print(f'XOR(0, 0) : {_xor_(0, 0)}')
print(f'XOR(0, 1) : {_xor_(0, 1)}')
print(f'XOR(1, 0) : {_xor_(1, 0)}')
print(f'XOR(1, 1) : {_xor_(1, 1)}')
```

### ■ AND のプロット

```python
def _and_plot_(x_data, y_data):

  w1 = 0.5
  w2 = 0.5
  b = -0.75

  x_plot_1, x_plot_2 = np.meshgrid(np.linspace(-0.5, 1.5, 100), np.linspace(-0.5, 1.5, 100))
  y_plot = w1 * x_plot_1 + w2 * x_plot_2 + b

  plt.figure(figsize=(5, 5))
  plt.contourf(x_plot_1, x_plot_2, y_plot, levels=[-10, 0, 10], colors=['#FFCCCC', '#CCCCFF'], alpha=0.5)
  plt.scatter(x_data[:, 0], x_data[:, 1], c=y_data, cmap='winter', edgecolors='k', s=100)

  plt.xlim(-0.1, 1.1)
  plt.ylim(-0.1, 1.1)
  plt.grid()
  plt.show()

x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_data = np.array([0, 0, 0, 1])

_and_plot_(x_data, y_data)
```
