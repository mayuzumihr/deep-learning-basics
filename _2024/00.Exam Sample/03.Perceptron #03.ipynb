### ■ パーセプトロン

- 活性化関数を使う

```python
import numpy as np
import matplotlib.pyplot as plt
```

### ■ 活性化関数

```python
#------------------------------------------------------------------------------
# Step 関数
#------------------------------------------------------------------------------
def step(x):
    y = np.where(x > 0, 1, 0)
    return y

#------------------------------------------------------------------------------
# Sigmoid 関数
#------------------------------------------------------------------------------
def sigmoid(x):
    y = 1 / (1 + np.exp(-x))
    return y

#------------------------------------------------------------------------------
# ReLU (Rectified Linear Unit) 関数
#------------------------------------------------------------------------------
def relu(x):
    y = np.maximum(0, x)
    return y

#------------------------------------------------------------------------------
# tanh (Hyperbolic tangent) 関数
#------------------------------------------------------------------------------
def tanh(x):
    y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    return y


#------------------------------------------------------------------------------
# Softmax 関数
#------------------------------------------------------------------------------
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # オーバーフロー対策
    y = exp_x / np.sum(exp_x)
    return y

#------------------------------------------------------------------------------
# Main
#------------------------------------------------------------------------------
x_plot = np.linspace(-5, 5, 100)
y_plot = step(x_plot)
plt.plot(x_plot, y_plot, label='Step Function')

x_plot = np.linspace(-5, 5, 100)
y_plot = sigmoid(x_plot)
plt.plot(x_plot, y_plot, label='Sigmoid Function')

x_plot = np.linspace(-5, 2, 100)
y_plot = relu(x_plot)
plt.plot(x_plot, y_plot, label='ReLU (Rectified Linear Unit) Function')

x_plot = np.linspace(-5, 5, 100)
y_plot = tanh(x_plot)
plt.plot(x_plot, y_plot, label='tanh (Hyperbolic tangent) Function')

plt.legend()
plt.grid()
plt.show()
```

### ■ 活性化関数の導関数

```python
#------------------------------------------------------------------------------
# Sigmoid 関数の導関数
#------------------------------------------------------------------------------
def sigmoid_deriv(x):
    y = sigmoid(x) * (1 - sigmoid(x))
    return y

#------------------------------------------------------------------------------
# ReLU 関数の導関数
#------------------------------------------------------------------------------
def relu_deriv(x):
    y = np.where(x > 0, 1, 0)
    return y

#------------------------------------------------------------------------------
# tanh 関数の導関数
#------------------------------------------------------------------------------
def tanh_deriv(x):
    y = 1 - np.tanh(x) ** 2
    return y

#------------------------------------------------------------------------------
# Softmax 関数の導関数
#------------------------------------------------------------------------------
def softmax_derivative(x):
    s = softmax(x)
    y = np.diag(s) - np.outer(s, s)
    return y

#------------------------------------------------------------------------------
# Main
#------------------------------------------------------------------------------
x_plot = np.linspace(-5, 5, 100)
y_plot = sigmoid_deriv(x_plot)
plt.plot(x_plot, y_plot, label='Derivative of Sigmoid Function')

x_plot = np.linspace(-5, 5, 100)
y_plot = relu_deriv(x_plot)
plt.plot(x_plot, y_plot, label='Derivative of ReLU Function')

x_plot = np.linspace(-5, 5, 100)
y_plot = tanh_deriv(x_plot)
plt.plot(x_plot, y_plot, label='Derivative of tanh Function')

plt.legend()
plt.grid()
plt.show()
```

### ■ パーセプトロンに活性化関数を適用する（Step 関数）

```python
#------------------------------------------------------------------------------
# AND Gate
#------------------------------------------------------------------------------
def _and_(x1, x2):

  x = [x1, x2]
  w = [0.5, 0.5]
  b = -0.75
  ans = np.dot(x, w) + b

  return step(ans)

#------------------------------------------------------------------------------
# OR Gate
#------------------------------------------------------------------------------
def _or_(x1, x2):

  x = [x1, x2]
  w = [0.5, 0.5]
  b = -0.25
  ans = np.dot(x, w) + b

  return step(ans)

#------------------------------------------------------------------------------
# NAND Gate
#------------------------------------------------------------------------------
def _nand_(x1, x2):

  x = [x1, x2]
  w = [-0.5, -0.5]
  b = 0.75
  ans = np.dot(x, w) + b

  return step(ans)

#------------------------------------------------------------------------------
# Main
#------------------------------------------------------------------------------
print(f'AND(0, 0) : {_and_(0, 0)}')
print(f'AND(0, 1) : {_and_(0, 1)}')
print(f'AND(1, 0) : {_and_(1, 0)}')
print(f'AND(1, 1) : {_and_(1, 1)}')
print('---')
print(f'OR(0, 0) : {_or_(0, 0)}')
print(f'OR(0, 1) : {_or_(0, 1)}')
print(f'OR(1, 0) : {_or_(1, 0)}')
print(f'OR(1, 1) : {_or_(1, 1)}')
print('---')
print(f'NAND(0, 0) : {_nand_(0, 0)}')
print(f'NAND(0, 1) : {_nand_(0, 1)}')
print(f'NAND(1, 0) : {_nand_(1, 0)}')
print(f'NAND(1, 1) : {_nand_(1, 1)}')
```
